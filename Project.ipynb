```python
word_list = set()
file = open('train','r')
for i in file:
    line = i.split(" ")[2:]
    tem1 = []
    for j in range(len(line)):
        if j % 2 == 0:
            tem1=tem1+[line[j]]

    for i in tem1:
        word_list.add(i)

file.close()

typ_wd_c_list = {}
typ_wd_c_list.setdefault("spam", {})
typ_wd_c_list.setdefault("ham", {})

for i in word_list:
    typ_wd_c_list["spam"].setdefault(i, 0)
    typ_wd_c_list["ham"].setdefault(i, 0)

file = open('train','r')
mtype_count = {}
total_mail = 0

for i in file:
    total_mail=total_mail+1
    line = i.split(" ")
    type = line[1]
    mtype_count.setdefault(type, 0)
    mtype_count[type]=mtype_count[type]+1

    for i in range(2, len(line), 2):
        typ_wd_c_list[type][line[i]]=typ_wd_c_list[type][line[i]]+int(line[i + 1])
    line = []
file.close()
totalham_w_fre = 0
for i,j in typ_wd_c_list["ham"].items():
    totalham_w_fre=totalham_w_fre+j

tspam_w_fre = 0
for i,j in typ_wd_c_list["spam"].items():
    tspam_w_fre=tspam_w_fre+j

prior = {}
for i,j in mtype_count.items():
    prior.setdefault(i, 0.0)
    prior[i] = float(j / total_mail)

probability_ew = {}
probability_ew.setdefault("ham", {})
probability_ew.setdefault("spam", {})

for i, j in typ_wd_c_list.items():
    for k, l in j.items():
        probability_ew[i].setdefault(k, 0)
        if (i in "spam"):
            probability_ew[i][k] =(l + 1) / float(tspam_w_fre + len(word_list))
        elif (i in "ham"):
            probability_ew[i][k] =(l + 1) / float(totalham_w_fre + len(word_list))

predicted_type = []
actual_type = []
email_id = []
pred_type = ""
testfile = open('test','r')
for i in testfile:
    line = i.split(" ")
    id = line[0]
    real_type = line[1]
    wordList = []
    for k in range(2, len(line)):
        if k % 2 == 0:
            wordList=wordList+[line[k]]

    ham_val =prior["ham"]
    spam_val =prior["spam"]
    for word in wordList:
        ham_val=ham_val*(probability_ew["ham"][word])
        spam_val=spam_val*(probability_ew["spam"][word])
    if spam_val > ham_val:
        pred_type = "spam"
    else:
        pred_type = "ham"

    actual_type.append(real_type)
    predicted_type.append(pred_type)
    email_id.append(id)
testfile.close()

true_positive = 0
true_negative = 0
false_positive = 0
false_negative = 0
for i in range(len(actual_type)):
    if (predicted_type[i] =="spam"):
        if (actual_type[i] =="spam"):
            true_positive=true_positive+1
        elif (actual_type[i] == "ham"):
            false_positive=false_positive+1

    elif (predicted_type[i] =="ham"):
        if (actual_type[i] =="ham"):
            true_negative=true_negative+1
        elif (actual_type[i] =="spam"):
            false_negative=false_negative+1

precision = (true_positive / float(true_positive + false_positive)) * 100
recall = (true_positive / float(true_positive + false_negative)) * 100
fmeasure = (2 * precision * recall) / (precision + recall)
print("       Recall = " + str(recall))
print("       Precision = " + str(precision))
print("       F1 Score (F score) = " + str(fmeasure))
print("\n\n")
print("   " + "ID" + "    " + "Actual_Type" + "   " + "Predicted_Type")
for qq in range(len(actual_type)):
    print(email_id[qq] + "     " + actual_type[qq] + "            " + predicted_type[qq])
```
